```python
# Customer Type
# Creating a new column named 'Customer Type' that will map 1 = 'Loyal Customer' and 2 = 'disloyal Customer'
# Create the new column by mapping the original values
air_2['Customer Type Num'] = air_2['Customer Type'].map({
    'Loyal Customer': 1,
    'disloyal Customer': 2
})
```

```python
order = ["neutral or dissatisfied", "satisfied"]

# Keep only what we need
plot_df = air_2[[
    "satisfaction",
    "Departure Delay in Minutes",
    "Arrival Delay in Minutes"
]].copy()

# Arrival delay often has NaNs in this dataset; either fill or drop
plot_df["Arrival Delay in Minutes"] = plot_df["Arrival Delay in Minutes"].fillna(0)

# Long format: one column for delay type, one for delay value
long_df = plot_df.melt(
    id_vars="satisfaction",
    value_vars=["Departure Delay in Minutes", "Arrival Delay in Minutes"],
    var_name="Delay Type",
    value_name="Delay (minutes)"
)

# Optional: prettier labels
long_df["Delay Type"] = long_df["Delay Type"].replace({
    "Departure Delay in Minutes": "Departure",
    "Arrival Delay in Minutes": "Arrival"
})

plt.figure(figsize=(10,4))
sns.boxplot(
    data=long_df,
    x="satisfaction",
    y="Delay (minutes)",
    hue="Delay Type",
    order=order,
    showfliers=False  # switch to True if you want to see extreme outliers
)

plt.title("Departure vs Arrival Delay by Satisfaction")
plt.xlabel("Satisfaction")
plt.ylabel("Delay (minutes)")
plt.legend(title="")
plt.tight_layout()
plt.show()
```

From this bar graph, it suggest that `Satisfied` delays are shorter than `neutral or dissatisfied` and the wait times goes up to 35 minutes while ` satisfied` is only about 20 minutes

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import RdYlGn5  # Red-Yellow-Green palette


# We calculate the average delay to make the story clear.
summary_data = air_2.groupby("satisfaction")["Departure Delay in Minutes"].mean().reset_index()
summary_data["satisfaction"] = summary_data["satisfaction"].astype(str) # Ensure categorical is string
source = ColumnDataSource(summary_data)

# Define the categories for the x-axis
categories = ["neutral or dissatisfied", "satisfied"]

# 2. Figure
p = figure(x_range=categories, height=400, width=700,
           title="Average Departure Delay by Passenger Satisfaction",
           toolbar_location=None, # Remove side toolbar for a cleaner look
           tools="")


# We map colors: Red/Orange for dissatisfied, Green for satisfied
p.vbar(x='satisfaction', top='Departure Delay in Minutes', width=0.5, source=source,
       line_color="white",
       fill_color=factor_cmap('satisfaction', palette=["#E74C3C", "#2ECC71"], factors=categories))


# This adds a tooltip so people can see the exact number when they hover
hover = HoverTool()
hover.tooltips = [
    ("Satisfaction", "@satisfaction"),
    ("Avg Delay", "@{Departure Delay in Minutes}{0.1f} min") # Format to 1 decimal place
]
p.add_tools(hover)

# Clean Up Visual
p.xgrid.grid_line_color = None   # Remove vertical grid lines
p.y_range.start = 0              # Start y-axis at 0
p.yaxis.axis_label = "Avg Delay (Minutes)"
p.xaxis.axis_label = ""          # No need for label, categories are obvious
p.outline_line_color = None      # Remove the border box

show(p)
```

Viewing the `Satisfaction` rating from the Airline, looking at the data, a majority of the people are satisfacted, however about the half to people are `unsatisfacted`

```python
# Viewing the
output_notebook()

# 1) Aggregate (pick the stat you want: mean / median / sum)
bar_df = (air_2
          .groupby("satisfaction", as_index=False)
          .agg(avg_flight_distance=("Flight Distance", "mean"),
               n=("Flight Distance", "size"))
         )

# (optional) force a nicer order if you only have these two categories
order = ["neutral or dissatisfied", "satisfied"]
bar_df["satisfaction"] = pd.Categorical(bar_df["satisfaction"], categories=order, ordered=True)
bar_df = bar_df.sort_values("satisfaction")

# 2) Bokeh source + plot
source = ColumnDataSource(bar_df)

p = figure(
    x_range=bar_df["satisfaction"].astype(str).tolist(),
    height=500,
    title="Average Flight Distance by Satisfaction",
    x_axis_label="Satisfaction",
    y_axis_label="Avg Flight Distance"
)

p.vbar(x="satisfaction", top="avg_flight_distance", width=0.7, source=source)
p.y_range.start = 0

# Nice hover
p.add_tools(HoverTool(tooltips=[
    ("Satisfaction", "@satisfaction"),
    ("Avg Distance", "@avg_flight_distance{0.0}"),
    ("Count", "@n")
]))

show(p)
```
